Questão1 - Realizar a média da metragem para cada um dos bairros. (Proposto pelo Paulo).

```python
bairros = dados['Bairro'].values # Nome dos bairros.
dict_media_metragem = {}

for bairro in bairros:
  media_metragem = dados.loc[dados['Bairro'] == bairro, 'Metragem'].mean() # Calcula a média da metragem de acordo com o nome do bairro.
  dict_media_metragem[bairro] = media_metragem # Adiciona a média ao dicionário.

dict_media_metragem
```

Questão3 - Explorar alguns gráficos na documentação e aplicar nas demais colunas do DF, assim como tentar colocar alguma conclusão.(Proposto pelo Thiago).

```python
moradia_bairros = dados['Bairro'].value_counts().head(5) # Filtra os 5 bairros com mais moradias disponíveis.
demais_bairros = dados['Bairro'].value_counts().sum() - moradia_bairros.sum() # Conta as moradias restantes

moradia_bairros['Outros bairros'] = demais_bairros

moradia_bairros.plot(kind='pie', autopct='%1.1f%%', fontsize=8, figsize=(7, 7)) # autopct='%1.1f%% mostra o valor em porcentagem no gráfico.
```

5 - Descobrir quais são os bairros que não tem nome de rua. (Proposto pela Vivian).

```python
# bairros que não possuem rua.
bairros_sem_rua = []

for bairro in dados['Bairro'].values:
  nao_nulos = dados.loc[dados['Bairro'] == bairro, 'Rua'].notnull().sum() # conta os não nulos.
  if nao_nulos == 0: # Se a quantidade de não nulos for 0, adiciona a lista.
    bairros_sem_rua.append(bairro)

bairros_sem_rua
```

3 - Preço do metro quadrado por bairro e plotar em um gráfico ideal.

Esse desafio pode ser feito por duas abordagens:

1 - Criar uma coluna com o preço do metro quadrado de cada imóvel, e fazer a média por bairros.

2 - Somar os valores dos imóveis e da área deles por bairros e fazer a média

```python
# 1: 
dados_vendas['Preco_m2'] = dados_vendas.Valor_anuncio_float / dados_vendas.Metragem

media_m2 = dados_vendas.groupby('Bairro')['Preco_m2'].mean().sort_values(ascending=False)
plt.figure(figsize=(15, 8))
media_m2[0:15].plot(kind='bar')
plt.title('Valor do metro quadrado dos 15 bairros mais caros de São Paulo')
plt.ylabel('Valor do m2')
plt.show()
```

```python
# 2: 
valores_bairro = dados_vendas.groupby('Bairro')['Valor_anuncio_float'].sum()
metragem_bairro = dados_vendas.groupby('Bairro')['Metragem'].sum()

valor_m2_2 = valores_bairro/metragem_bairro
valor_m2_2 = valor_m2_2.sort_values(ascending=False)
plt.figure(figsize=(15, 8))
valor_m2_2[:15].plot(kind='bar')

plt.title('Valor do metro quadrado dos 15 bairros mais caros de São Paulo')
plt.ylabel('Valor do m2')
plt.show()
```

 O resultado foi diferente para cada forma de calculo.

 Variação dependente do tipo de busca.

 Primeiro grafico foi obtido a média do valor de metro quadrado por imóvel para cada bairro.

 Segundo gráfico foi obtido o valor do metro quadrado por bairro, e não a média do valor de metro quadrado por imóvel.

 É notável que nos 2 gráficos o bairro Vila Guaca possui um valor muito acima dos demais. Portanto, filtrando os imóveis deste bairro abaixo, notamos que existe apenas um imóvel para ele, com 10 metros quadrados de área, 3 quartos e 2 banheiros anunciado por 800 mil reais. Claramente está errado esse valor de 10 metros quadrados em um imóvel com 3 quartos e 2 banheiros. Por isso a distorção do valor bem acima por Mts2 apresentado no gráfico.

5 - Pesquisar um visualização para analisar os quartis, mediana e outliers

```python
plt.figure(figsize=(15, 8))
sns.boxplot(data=dados_vendas, y = 'Valor_anuncio_float')
plt.title('Boxplot Valor dos imóveis')
plt.ylabel('Valor dos imóveis')
plt.yticks([0, 200000, 500000, 1000000, 2000000, 5000000, 10000000])
plt.show()
```

 Devido a grande quantidade de outliers, a visualização do primeiro e terceiro quartil ficam prejudicados, assim como a mediana. Dessa forma, será limitado o eixo y em 8 milhões, para facilitar a visualização.

```python
plt.figure(figsize=(15, 8))
sns.boxplot(data=dados_vendas, y = 'Valor_anuncio_float')
plt.title('Boxplot Valor dos imóveis')
plt.ylabel('Valor dos imóveis')
plt.yticks([0, 200000, 500000, 1000000, 2000000, 5000000, 7000000])
plt.ylim((0,8000000))
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.boxplot(data=dados_vendas[['Quartos', 'Banheiros', 'Vagas']])
plt.title('Distribuição de número de quartos, banheiros e vagas por imóvel')
plt.ylabel('Quantidade')
plt.show()
```

 Da mesma forma que no valor dos imóveis, a grande quantidade de outliers prejudica a visualização. Mas os valores detalhados podem ser vistos a seguir e comparados nos gráficos. O valor da mediana corresponde a linha 50%.
